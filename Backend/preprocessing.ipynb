```python
import pandas as pd
```

```python
df = pd.read_csv('../Dataset/Word_University_Rank_2020.csv')
df.head()
```

**Removing percentage sign from the data**


```python
df['International_Students'] = df['International_Students'].str.strip('%')
df['Percentage_Female'] = df['Percentage_Female'].str.strip('%') 
df['Percentage_Male'] = df['Percentage_Male'].str.strip('%')
```

```python
df['International_Students']=pd.to_numeric(df['International_Students'], errors='coerce') # two methods to convert to float
df['Percentage_Female']= df['Percentage_Female'].apply(lambda x:float(x))
df['Percentage_Male']= df['Percentage_Male'].apply(lambda x:float(x))
```

```python
df.head()
```

**Convert data to percentages**

```python
df[df.columns[5:]] /=100
```
